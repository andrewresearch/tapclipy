## Vocabulary

Vocabulary is a query that returns the stats on the vocabulary used, It groups them by unique words and how many times they were used.

### Example:

```python
# Set our query type to Vocabulary
query = tap.query('vocabulary')

# no params needed

# pass in some test data
string = "This is the first great happy blue angry cold sentence. This is the second fantastic sentence."

# query the api
strResult = tap.analyse_text(query, string)

uniqueWords = strResult['data']['vocabulary']['analytics']['unique']

output = "\n"

for term in strResult['data']['vocabulary']['analytics']['terms']:
    output += "{0} : {1}\n".format(term['term'], term['count'])

# Print Result
print("-" * 40)
print("Vocabulary:")
print("-" * 40)


print("Input Text:\n\n", string)

print("\n")
print("Unique Words:\n", uniqueWords)

print("\n")
print("Word Occurences:\n", output)

print("\n")
print("Raw Result:\n\n", json.dumps(strResult, indent=2))
```
